```python
import pandas as pd
import utils
```

```python
df_combined_sorted = pd.read_csv("./data/scandals_20200101to20241113_symbol_added_v4.csv")
```

```python
df_length = df_combined_sorted.shape[0]
```

```python
for index, row in df_combined_sorted.iterrows():
    
    if row["scandal_of_company"] != "Y":
        continue

    if row["related_company_name"] == "-":
        keyword = row["company_name"]
    else:
        keyword = row["related_company_name"]

    code = utils.get_stock_code(keyword)
    print(f"keyeord: {keyword}, code: {code}   {index}/{df_length}")

    df_combined_sorted.loc[index, "market_code"] = code
    
```

```python
df_combined_sorted_filled = df_combined_sorted.fillna("-")
df_combined_sorted_filled
```

```python
df_combined_sorted_filled.to_csv("./data/scandals_20200101to20241113_symbol_added_v5.csv")
```
